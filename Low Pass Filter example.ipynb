# Low Pass Filter(LPF) example
* Building of box-type low pass filter using Python
* zero-padding


```python
import numpy as np
import numpy.fft
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
N = 10
x = np.ones(N)
X = fft.fft(x)
f_x = np.linspace(0,2*np.pi,N)
#ff_x = fft.fftfreq(N, 1)

N_zeros = 100
y = np.array([1]*N+[0]*N_zeros)
Y = fft.fft(y)
f_y = np.linspace(0,2*np.pi,(N+N_zeros))
#ff_y = fft.fftfreq(N+N_zeros, 1)

N_z = 100
z = np.zeros(N_z)
z[:N_z//2] = np.exp(-np.arange(N_z//2)/10)
Z = fft.fft(z)
f_z = np.linspace(0,2*np.pi,N_z)
```

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(111)
ax.plot(f_x, np.abs(X),'ro-')
ax.plot(f_y, np.abs(Y),'b-')
ax.plot(f_z, np.abs(Z),'g-')
ax.set_xlim([-1,7]);ax.set_ylim([-1, 11])
ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(111)
ax.semilogy(f_y, np.abs(Y))
ax.semilogy(f_z, np.abs(Z))
```
